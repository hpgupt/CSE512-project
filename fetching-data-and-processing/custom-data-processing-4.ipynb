```python
text_files_directory = '/content/lyric-mood-classification/data/lyrics/txt'
import os

for index, row in df_genres.iterrows():
    # Get the filename from the 'lyrics_filename' column
    filename = str(row['lyrics_filename'])+'.txt'

    # Construct the full path to the text file
    file_path = os.path.join(text_files_directory, filename)

    # Check if the file exists
    if os.path.exists(file_path):
        # Read the contents of the text file
        with open(file_path, 'r') as file:
            file_contents = file.read()

        # Assign the file contents to the 'lyrics' column for the current row
        df_genres.at[index, 'lyrics'] = file_contents
```
